```python
#Filtering out the unnecessary columns
tree_census_subset = tree_census[['tree_id', 'created_at', 'tree_dbh', 'stump_diam',
       'curb_loc', 'status', 'health', 'spc_latin', 'steward',
     'sidewalk', 'user_type', 'problems', 'root_stone',
       'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
       'brch_light', 'brch_shoe', 'brch_other']]
tree_census_subset
```

```python
#Representing the data graphically/Visually

tree_census_subset.hist(bins = 60, figsize = (20,10)) #figsize was used just to increase size of the graph
```
